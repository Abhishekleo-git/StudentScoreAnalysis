# Drop Unnamed column

```python
df = df.drop("Unnamed: 0", axis = 1)
```

```python
print(df.head())
```

```python
df["IsFirstChild"] = df["IsFirstChild"].str.replace("no","nopes")
print(df.head())
```

```python
plt.figure(figsize= (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

# from above chart we analysed that no. of females is more than no. of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb, annot = True)
plt.title("Parent's education V/S Scores")
```

# from above heatmap we can observe that there is a great impact of parent's education on scores of students
